###logistic regression for multiclass

```python
#make a prediction with a multiclass logitic regression model
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
#define dataset
x,y=make_classification(n_samples=1000,n_features=10,n_informative=5,n_redundant=5,n_classes=3,random_state=1)
```

```python
x
```

```python
y
```

```python
#define multinomial logistic regresssion model for multiclass classification
model=LogisticRegression(multi_class='multinomial',solver='lbfgs')
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
#training model
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
model.predict_proba(x_test)#jiski probality jada hogi wo output hoga
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
