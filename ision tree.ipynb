###decision tree classifier implementation


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
y=dataset.target
#df.iloc: This refers to integer-location based indexing in pandas. It allows you to select rows and columns by their position (index).

#: (before the comma): This selects all rows of the DataFrame.

#:-1 (after the comma): This selects all columns except the last one. The -1 is a slice notation, which means "up to but not including the last element
```

```python
x,y
```

```python
#traimtest spilt
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#post prunining
treeclassifier=DecisionTreeClassifier()#use default parameters
treeclassifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
#post prunining with maxdepth =2
treeclassifier=DecisionTreeClassifier(max_depth=2) 
treeclassifier.fit(x_train,y_train)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
##prediction
y_pred=treeclassifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```

##pre prunning and hyer parameter tuning


```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameters={ 
'criterion':['gini',"entropy",'log_loss'] ,
'splitter':['best','random'],
'max_depth':[1,2,3,4,5],
'max_features':['auto','sqrt','log2']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeclassifier=DecisionTreeClassifier()
clf=GridSearchCV(treeclassifier,param_grid=parameters,cv=5,scoring='accuracy')
```

```python
clf.fit(x_train,y_train)
```

```python
clf.best_params_
```

```python
##prediction
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
score=accuracy_score(y_pred,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```
